```python
df2 = pd.read_csv(r'C:\Users\Samir\Documents\ML\DataSets\reviews.csv')
```

```python
merged_df = pd.merge(right = df, left=df2[['listing_id', 'comments']], left_on='listing_id', right_on='id', how='inner')
```

```python
#these are top 20 neighbourhoods
X = merged_df['neighbourhood'].value_counts()[:50]
X
```

```python
neighbourhood_names = X.index
neighbourhood_names
```

```python
top20_neighbourhood = merged_df[merged_df['neighbourhood'].isin(neighbourhood_names)]
```

```python
len(top20_neighbourhood)
```

```python
top20_neighbourhood.groupby('neighbourhood')['price'].mean()
```

```python
top20_neighbourhood['comments'] = top20_neighbourhood['comments'].astype(str).fillna('')

neighborhood_descriptions = top20_neighbourhood.groupby('neighbourhood')['comments'].apply(' '.join).reset_index()
```

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

additional_stopwords = {'seattle', 'neighborhood', 'place', 'area','the','was','in','to','location','stay','host','house','home'}
stopwords = ENGLISH_STOP_WORDS.union(additional_stopwords)
```

```python
import re
def clean_text(text):
    text = text.lower() 
    text = re.sub(r'\d+', '', text)  
    text = re.sub(r'[^\w\s]', '', text)  
    words = text.split()
    words = [word for word in words if word not in stopwords]  
    return ' '.join(words)

neighborhood_descriptions['cleaned_description'] = neighborhood_descriptions['comments'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=5)

neighborhood_descriptions['top_words'] = neighborhood_descriptions['cleaned_description'].apply(
    lambda x: vectorizer.fit([x]).get_feature_names_out()
)

for index, row in neighborhood_descriptions.iterrows():
    print(f"{row['neighbourhood']}: {', '.join(row['top_words'])}")
```
